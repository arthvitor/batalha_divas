# Análise - Batalha de músicas
Etapa de análise exploratória de dados e adequação dos dados para serem recebidos no Front-end

## Importando bibliotecas e base

```python
# importando libs
import pandas as pd
pd.set_option('display.max_columns', None) # Aumenta o número máximo de colunas printadas
```

```python
# dando entrada no Dataframe
df = pd.read_csv(r'data\Top_10_Tours_Bey_Tay.csv', sep=',')
df.head(1)
```

```python
# Tratando a coluna Loudness
df['Loudness'] = df['Loudness'] * - 1
```

## AED das músicas por métrica

```python
# Selecionando somente músicas da Beyoncé
musica_bey = df.query(' Artist == "Beyoncé" ')
musica_bey
```

```python
# Selecionando somente músicas da Taylor Swift
musica_taylor = df.query(' Artist == "Taylor Swift" ')
musica_taylor
```

```python
# Resumo estatístisco das músicas da Taylor
resumo = musica_taylor.describe()
print(resumo)
```

```python
# verificando música vencedora por métrica
musica_taylor.sort_values(by=['Loudness'], ascending=False).head(1)
```

```python
# verificando música vencedora por métrica - Bey
musica_bey.sort_values(by=['Loudness'], ascending=False).head(1)
```

```python
# verificando música vencedora por métrica - GERAL
df.sort_values(by=['danceability'], ascending=False).head(1)
```

# Tratamento

```python
# Criando arquivo com dados para gráfico no front-end
df_contat = df.groupby('Artist')[['Popularity', 'danceability', 'Loudness', 'valence', 'Energia']].mean().reset_index().round(3)
df_contat.to_csv('Top_10_Tours_Bey_Tay_TRATADO1.csv', sep=',')
```
